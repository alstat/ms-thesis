### i. Monte Carlo Simulation
The following codes simulates the monte carlo method by approximating the area under the Gaussian distribution between -1.96 and 1.96.

```julia
using StatsBase
using Distributions
using PyPlot

col, bgcol, grdcol, lty, ticksize = ("black", "white", "#848482", "dotted", 8)
```

#### My Plot Function

```julia
function myplot(x, y = Union{}; xlab = "", ylab = "", color = col, marker = "", markersize = 5, 
    ptype = "xyplot", where = "post", width = 1, bins = 30)
    if (y == Union{}) & (ptype != "hist")
        y = x
        ylen = y |> length
        x = collect(1:ylen)
        
        ylim_margin = -(y |> maximum, y |> minimum) * .05
        xlim_margin = -(x |> maximum, x |> minimum) * .05
    elseif (y == Union{}) & (ptype == "hist")
        y = x
        ylim_margin = -(y |> maximum, y |> minimum) * .05
        xlim_margin = -(x |> maximum, x |> minimum) * .05
    else
        ylim_margin = -(y |> maximum, y |> minimum) * .05
        xlim_margin = -(x |> maximum, x |> minimum) * .05
    end
   
    f, ax = subplots(figsize = (5, 5))
    
    if ptype == "step"
        ax[:step](x, y, where = where, color = color, marker = marker, markersize = markersize)
    elseif ptype == "xyplot"
        ax[:plot](x, y, color = color, marker = marker, markersize = markersize)
    elseif ptype == "bar"
        ax[:bar](x, y, width, color = color)
    elseif ptype == "hist"
        ax[:hist](x, bins, edgecolor = "white", facecolor = "black")
    end
    
    ax[:set_xlabel](xlab)
    ax[:set_ylabel](ylab)
    ax[:grid]("on", which = "major", color = grdcol, linestyle = lty)
    ax[:set_axis_bgcolor](bgcol)
    ax[:set_ylim](((y |> minimum) - ylim_margin, (y |> maximum) + ylim_margin))
    ax[:set_xlim](((x |> minimum) - xlim_margin, (x |> maximum) + xlim_margin))
    ax[:tick_params](axis = "both", which = "major", labelsize = ticksize)
    ax[:tick_params](axis = "both", which = "minor", labelsize = ticksize)
    ax[:set_axisbelow]("on")
end
```

```julia
srand(123)
draws = collect(1000:100:100000)
area = draws |> length |> zeros

for i in 1:(draws |> length)
    samples = rand(Normal(), draws[i])
    area[i] = sum((samples .> -1.96) & (samples .< 1.96)) / (samples |> length)
end
```

```julia
myplot(area, xlab = "Draws", ylab = "Area")
```

### ii. Metropolis-Hasting

The following codes illustrates the Metropolis-Hasting algorithm by using uniform distribution as the proposal distribution, with target distribution given by univariate Cauchy model.

```julia
r = 100000
x = r |> zeros
x[1] = 30

for i in 1:r - 1
    propose = x[i] + rand(Uniform(-1, 1))
    accept = rand(Uniform()) < pdf(Cauchy(), propose) / pdf(Cauchy(), x[i])

    if accept == true
        x[i + 1] = propose
    else
        x[i + 1] = x[i]
    end
end
```

```julia
myplot(x[1000:end], xlab = "Iterations", ylab = "Mixing")
```

Another example of Metropolis-Hasting algorithm for sampling on bivariate Cauchy distribution

```julia
function BvCauchy(x,  = [0; 0],  = 1)
  (1 / (2pi)) * ( / ((x[1] - [1])^2 + (x[2] - [2])^2 + ^2)^1.5)
end

x = zeros((r, 2))
x[1, :] = [-100; 100]

for i in 1:r - 1
    propose = x[i, :]' + rand(Uniform(-5, 5), 2)
    accept = rand(Uniform(), 1)[1] < (BvCauchy(propose) / BvCauchy(x[i, :]))

    if accept == true
        x[i + 1, :] = propose
    else
        x[i + 1, :] = x[i, :]
    end
end
```

```julia
function bv_norm(x, 1 = 10, 2 = -10, 1 = 1.5, 2 = 1.35,  = .5)
    1 / (2*pi*1*2*sqrt(1-^2)) * exp(-(1 / (2*(1 - ^2))) * ((((x[1, 1] - 1)^2) / (1^2)) +
    (((x[2, 1] - 2)^2) / 2^2) - ((2**(x[1, 1] - 1)*(x[2, 1] - 2)) / (1*2))))
end

x = zeros((r, 2))
x[1, :] = [0 0]

for i in 1:r - 1
  proposal = x[i, :]' + rand(Uniform(-5, 5), 2)
  accept = rand(Uniform(), 1)[1] < (bv_norm(proposal) / bv_norm(x[i, :]'))

  if accept == true
    x[i + 1, :] = proposal
  else
    x[i + 1, :] = x[i, :]
  end

end
```

## Hamiltonian Monte Carlo Algorithm

The following algorithm is an implementation of the Hamiltonian Monte Carlo

```julia
 = .3
 = 20
 = [10; -10]                                       # Define the mean vector of target distribution
 = [1.5^2 (1.5)(1.35).5; (1.5)(1.35).5 1.35^2]     # Define the variance-covariance matrix of target distribution

U(x, , ) = (x - )' * inv() * (x - )               # Potential Energy (Target Distribution: Gaussian)
dU(x, , ) = inv() * (x - )                         # Gradient of U

K(p) = (p'p) / 2                                    # Define Kinetic Energy (Proposal: Standard Gaussian)
dK(p) = p                                           # Gradient of K

H(x, p, , ) = U(x, , ) + K(p)                         # Define Hamiltonian

x = zeros(2, r); x[:, 1] = [0; 0];                  # Initialize Position

for n = 1:r-1
    xNew = x[:, n]
    p = rand(Normal(), size(xNew))                  # Initialize Momentum
    oldE = H(xNew, p, , )                            # Compute the current Total Energy
    
    for t = 1:                                     # Perform leap frog method
        p = p - ( / 2)dU(xNew, , )          
        xNew = xNew + ()dK(p)
        p = p - ( / 2)dU(xNew, , )
    end
    
    newE = H(xNew, p, , )                            # Compute the new Total Energy
    dE = newE - oldE                                # Compute changes in Energy
    
    if dE[1] < 0
        x[:, n + 1] = xNew
    elseif rand(Uniform()) < exp(-dE[1])
        x[:, n + 1] = xNew
    else
        x[:, n + 1] = x[:, n]
    end
    
end
```

```julia
myplot(x[1, 1:2000]', x[2, 1:2000]', xlab = "x1", ylab = "x2")
myplot(autocor(x[1, 1:200]'), xlab = "Lags", ylab = "Autocorrelation", ptype = "bar", width = .5)
myplot(autocor(x[2, 1:200]'), xlab = "Lags", ylab = "Autocorrelation", ptype = "bar", width = .5)
```

```julia
myplot(x[1, :]', x[2, :]', marker = "o", color = "white", xlab = "x1", ylab = "x2")
myplot(autocor(x[1, :]'), xlab = "Lags", ylab = "Autocorrelation", ptype = "bar", width = .5)
#myplot(autocor(x[2, ]'), xlab = "Lags", ylab = "Autocorrelation", ptype = "bar", width = .5)
```

```julia
delta = .3
nSamples = r
L = 20
 = [10; -10]                                       # Define the mean vector of target distribution
 = [1.5^2 (1.5)(1.35).5; (1.5)(1.35).5 1.35^2]     # Define the variance-covariance matrix of target distribution

U(x, , ) = (x - )' * inv() * (x - )               # Potential Energy (Target Distribution: Gaussian)
dU(x, , ) = inv() * (x - )                         # Gradient of U

K(p) = (p'p) / 2                                    # Define Kinetic Energy (Proposal: Standard Gaussian)
dK(p) = p                                           # Gradient of K

H(x, p, , ) = U(x, , ) + K(p)                         # Define Hamiltonian

x = zeros(2, nSamples) 
x0 = [0; 0]                  # Initialize Position
x[:, 1] = x0

n = 1
while n < nSamples
    n = n + 1;
 
    # SAMPLE RANDOM MOMENTUM
    p0 = rand(Normal(), size(x[:, 1]))  
 
    # SIMULATE HAMILTONIAN DYNAMICS
    # FIRST 1/2 STEP OF MOMENTUM
    pStar = p0 - (delta/2)*dU(x[:, n-1], , )
 
    # FIRST FULL STEP FOR POSITION/SAMPLE
    xStar = x[:,n-1] + delta*pStar
 
    # FULL STEPS
    for  jL = 1:L-1
        # MOMENTUM
        pStar = pStar - delta*dU(xStar, , );
        # POSITION/SAMPLE
        xStar = xStar + delta*pStar;
    end
 
    # LAST HALP STEP
    pStar = pStar - (delta/2)*dU(xStar, , );
 
    # COULD NEGATE MOMENTUM HERE TO LEAVE
    # THE PROPOSAL DISTRIBUTION SYMMETRIC.
    # HOWEVER WE THROW THIS AWAY FOR NEXT
    # SAMPLE, SO IT DOESN'T MATTER
 
    # EVALUATE ENERGIES AT
    # START AND END OF TRAJECTORY
    U0 = U(x[:, n - 1], , );
    UStar = U(xStar, , );
 
    K0 = K(p0);
    KStar = K(pStar);
 
    # ACCEPTANCE/REJECTION CRITERION
    alpha = minimum([1; exp((U0 + K0) - (UStar + KStar))])
 
    u = rand(Uniform())
    if u < alpha
        x[:, n] = xStar
    else
        x[:, n] = x[:, n - 1]
    end
end
```

Derive the log likelihood function from the model

```julia
function loglikelihood(0, 1, ; y = y, x = x)
    
     = 0 + 1*x
    
    loglike = zeros(length())
    for i in 1:length()
        loglike[i] = logpdf(Normal([i], ), y[i])
    end
    
    return loglike |> sum
end

function loglike_profile(x, )
    
    prof = zeros(length(x))
    for i in 1:length(x)
        prof[i] = loglikelihood(x[i], [1], [2]; y = [3], x = [4])
    end
    
    return prof
end
```

Define the prior

```julia
function prior()
    0_prior = logpdf(Normal(0, 5), [1]) #3
    1_prior = logpdf(Uniform(0, 10), [2]) #10
    _prior = logpdf(Uniform(0, 30), [3]) #30
    
    return +(0_prior, 1_prior, _prior)
end
```

Define the posterior using the fact that

\begin{equation}
\mathbb{P}(\boldsymbol{\theta}|\mathbf{X}) = \frac{\mathbb{P}(\mathbf{X}|\boldsymbol{\theta})\mathbb{P}(\boldsymbol{\theta})}{\mathbb{P}(\mathbf{X})}
\end{equation}

So that the log pdf is

\begin{equation}
\mathbb{P}(\boldsymbol{\theta}|\mathbf{X}) = \log\mathbb{P}(\mathbf{X}|\boldsymbol{\theta}) + \log\mathbb{P}(\boldsymbol{\theta}) - \log\mathbb{P}(\mathbf{X})
\end{equation}

Such that the unnormalize log pdf is given by

\begin{equation}
\mathbb{P}(\boldsymbol{\theta}|\mathbf{X}) = \log\mathbb{P}(\mathbf{X}|\boldsymbol{\theta}) + \log\mathbb{P}(\boldsymbol{\theta})
\end{equation}

```julia
function posterior(; y = y, x = x)
    return +(loglikelihood([1], [2], [3]; y = y, x = x), prior())
end
```

Metropolis-Hasting Algorithm

```julia
function proposal()
    random = zeros(3)
    
    j = 1
    for i in [.1 .5 .3]
        random[j] = rand(Normal([j], i), 1)[1]
        j += 1
    end
    
    return random
end

function run_metropolis_MCMC(start::Array{Int64,1}, iter::Int64) 
    chain = zeros(iter + 1, 3)
    chain[1, :] = start
    
    for i in 1:iter
        propose = proposal(chain[i, :])
        
        probab = exp(posterior(propose; y = y, x = x) - posterior(chain[i, :]; y = y, x = x))
        
        if rand(Uniform()) < probab
            chain[i + 1, :] = propose
        else
            chain[i + 1, :] = chain[i, :]
        end
    end
    
    return chain 
end
```

#### Define the log Prior

```julia
function logprior(;  = ,  = ) 
    w0_p = logpdf(Normal([1], ), [1])
    w1_p = logpdf(Normal([2], ), [2])
    
    return +(w0_p, w1_p)
end
```

#### Define the log Likelihood

```julia
function loglike(;  = , x = x, y = y)
     = [1] + [2] * x
    
    likhood = zeros(x |> length)
    for i in 1:length()
        likhood[i] = logpdf(Normal([i], ), y[i])
    end
    
    return likhood |> sum
end
```

#### Define the log Posterior

```julia
function logpost()
    return +(loglike(,  = , x = x, y = y), logprior(,  = ,  = ))
end
```

#### Define the MCMC

Proposal function

```julia
s = [1, 1]
function proposal() 
    random = zeros( |> length)
    
    for i in 1:( |> length)
        random[i] = rand(Normal([i], s[i]), 1)[1]
    end
    return random
end
```

```julia
function mh(_start, max_iter)
    chain = zeros(max_iter + 1, _start |> length)
    chain[1, :] = _start
    
    for i in 1:max_iter
        propose = proposal(chain[i, :])
        probab = exp(logpost(propose) - logpost(chain[i, :]))
        
        if rand(Uniform(), 1)[1] < probab
            chain[i + 1, :] = propose
        else
            chain[i + 1, :] = chain[i, :]
        end
    end
    
    return chain
end
```
